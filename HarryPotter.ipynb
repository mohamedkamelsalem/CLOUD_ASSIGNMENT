```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("books.csv")
df
```

```python
harry_potter = df[df["title"].str.contains("Harry Potter", case=False, na=False)]
print(harry_potter)
```

```python
cdf = harry_potter[
    [
        "book_id",
        "books_count",
        "title",
        "average_rating",
        "ratings_count",
    ]
]
cdf.head(10)
```

```python
harry_potter_selling = cdf.sort_values(by="ratings_count", ascending=False)
harry_potter_selling
```

```python
average_rating_harry_potter = harry_potter["average_rating"].mean()
average_rating_harry_potter
```
